```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
df_2023 = pd.read_csv('../datasets/flights_2023.csv', low_memory=False)
df_2024 = pd.read_csv('../datasets/flights_2024.csv', low_memory=False)

print("2023 shape:", df_2023.shape)
print("2024 shape:", df_2024.shape)
print("\n2023 columns:", df_2023.columns.tolist())
print("\n2024 columns:", df_2024.columns.tolist())
```

```python
display(df_2023.head())
print(df_2023.info())

print("\n2023 missing values:")
print(df_2023.isnull().sum())

df_2023.describe()
```

```python
delay_cols = [col for col in df_2023.columns if 'delay' in col.lower()]
cancel_cols = [col for col in df_2023.columns if 'cancel' in col.lower()]
airport_cols = [col for col in df_2023.columns if col in ['origin', 'dest', 'origin_airport', 'dest_airport']]

print("Delay columns:", delay_cols)
print("Cancellation columns:", cancel_cols)
print("Airport columns:", airport_cols)
```

```python
if 'origin_airport' in df_2023.columns:
    top_airports = df_2023['origin_airport'].value_counts().head(10)
    top_airports.plot(kind='bar', title='Top 10 Origin Airports (2023)')
    plt.xticks(rotation=45)
    plt.show()
```

```python
if 'arr_delay' in df_2023.columns:
    plt.figure(figsize=(12,4))
    
    plt.subplot(1,2,1)
    df_2023['arr_delay'].hist(bins=50)
    plt.title('Arrival Delay Distribution (2023)')
    plt.xlabel('Minutes')
    
    plt.subplot(1,2,2)
    (df_2023['arr_delay'] > 15).value_counts().plot(kind='bar')
    plt.title('% Delayed >15min')
    plt.xticks([0,1], ['On Time', 'Delayed'], rotation=0)
    plt.ylabel('Count')
    
    plt.tight_layout()
    plt.show()
```

```python
df_combined = pd.concat([df_2023.assign(year=2023), df_2024.assign(year=2024)])

if 'arr_delay' in df_combined.columns:
    yearly_delay = df_combined.groupby('year')['arr_delay'].agg(['mean', 'count']).round(1)
    print("Average arrival delay by year:")
    print(yearly_delay)
```

```python
print("=== RECOMMENDED MAPPINGS ===")
print("""
flight_date → flight_date (pd.to_datetime().date())
reporting_airline → reporting_airline
flight_num_reporting_airline → flight_num_reporting_airline (int)
origin → origin
dest → dest
crs_dep_time → crs_dep_time (HHMM → DateTime)
dep_time → dep_time (HHMM → DateTime, null-safe)
crs_arr_time → crs_arr_time (HHMM → DateTime) 
arr_time → arr_time (HHMM → DateTime, null-safe)
dep_delay → dep_delay_minutes (int, null-safe)
dep_delay_minutes → dep_delay_minutes (int, null-safe)  
arr_delay → arr_delay_minutes (int, null-safe)
arr_delay_minutes → arr_delay_minutes (int, null-safe)
dep_del_15 → dep_del_15 (int)
arr_del_15 → arr_del_15 (int)
cancelled → cancelled (int)
cancellation_code → cancellation_code (if != 'NA')
diverted → diverted (int)
distance → distance (int)

DELAY CAUSES
carrier_delay → carrier_delay (int, null-safe)
weather_delay → weather_delay (int, null-safe)
nas_delay → nas_delay (int, null-safe)
security_delay → security_delay (int, null-safe)  
late_aircraft_delay → late_aircraft_delay (int, null-safe)
""")

# Quick validation
essential_cols = ['flight_date', 'reporting_airline', 'origin', 'dest', 
                 'dep_delay_minutes', 'arr_delay_minutes', 'cancelled']
missing_cols = [col for col in essential_cols if col not in df_2023.columns]
print(f"\nAll essential columns present: {len(missing_cols) == 0}")

print(f"\nSample LAX flights: {len(df_2023[df_2023['origin'] == 'LAX']):,}")
print(f"Sample delay distribution: {df_2023['arr_delay_minutes'].describe().round(1)}")
print(f"Cancellation rate: {df_2023['cancelled'].mean():.1%}")
```

```python
print("=== DATA QUALITY SUMMARY ===")
print(f"Total flights: {len(df_2023) + len(df_2024):,} (6.8M + 7.1M)")
print(f"Time range: {df_2023['flight_date'].min()} → {df_2024['flight_date'].max()}")
print(f"Unique airlines: {df_2023['reporting_airline'].nunique()}")
print(f"Unique airports: {df_2023['origin'].nunique()} origins, {df_2023['dest'].nunique()} dests")

# Delay stats
if 'arr_delay_minutes' in df_2023:
    print(f"\nArrival delays: {df_2023['arr_delay_minutes'].mean():.1f}min avg")
    print(f"   Delayed >15min: {df_2023['arr_del_15'].mean():.1%}")
    print(f"   Cancelled: {df_2023['cancelled'].mean():.2%}")

print("\nSchema-ready: All essential columns present")
print("Ready for analytics: Delay causes + cancellation codes included")
```
